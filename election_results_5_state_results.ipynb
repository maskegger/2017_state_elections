```python
import pandas
#df_1 = pandas.read_excel("http://eci.nic.in/eci_main/archiveofge2014/34%20Details%20of%20Assembly%20segment%20of%20PC.xls")
#you can use the command above to download and process the file, I prefer to process the files after they've\
#been downloaded, so that I don't have to redownload every single time I analyse it

df_1 = pandas.read_excel("34 Details of Assembly segment of PC.xls")
df_1.head()
```

```python
df_1.columns = df_1.iloc[0]
#df_2 = df_1.ix[1:]
df_1.head()
```

```python
df_2 = df_1.ix[1:]
df_2.head()
```

```python
df_2.reset_index(drop=True, inplace=True)
df_2.head()
```

```python
df_2.dtypes
```

```python
df_2['VOTESSECURED EVM'] = df_2['VOTESSECURED EVM'].astype(float)
df_2.head()
```

```python
df_2["ac_vote_max"] = df_2.groupby("AC NAME")["VOTESSECURED EVM"].transform(max)
df_2.head()
```

```python
df_3 = df_2.loc[df_2["VOTESSECURED EVM"] == df_2["ac_vote_max"]]
df_3.head()
```

```python
df_3["State-UT Code & Name"].unique()
```

```python
df_4 = df_3[df_3['State-UT Code & Name'].isin(['Goa','Manipur','Punjab','Uttar Pradesh','Uttarakhand'])]
df_4.head()
```

```python
df_4.groupby(['State-UT Code & Name','PARTY'])['PARTY'].count()
```

```python
df_uttarakhand2 = df_4[df_4['State-UT Code & Name']== 'Uttarakhand']
df_uttarakhand2.to_csv('uttarakhand_2014B.csv')
```

```python
df_goa2 = df_4[df_4['State-UT Code & Name']== 'Goa']
df_goa2.to_csv('goa_2014B.csv')
df_manipur2 = df_4[df_4['State-UT Code & Name']== 'Manipur']
df_manipur2.to_csv('manipur_2014B.csv')
df_punjab2 = df_4[df_4['State-UT Code & Name']== 'Punjab']
df_punjab2.to_csv('punjab_2014B.csv')
df_uttar_pradesh2 = df_4[df_4['State-UT Code & Name']== 'Uttar Pradesh']
df_uttar_pradesh2.to_csv('uttar_pradesh_2014B.csv')
```
